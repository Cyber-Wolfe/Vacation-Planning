## Deliverable 2. Create a Customer Travel Destinations Map.

```python
#Dependencies
import pandas as pd
import requests
import gmaps
import numpy as np
#Keys
from config import g_key
gmaps.configure(api_key=g_key)
```

```python
city_data_df = pd.read_csv("../Weather_Databases/WeatherPy_Database.csv")
city_data_df = city_data_df.drop(["City"], axis = 1)
city_data_df = city_data_df.rename(columns= {"City.1": "City"})
city_data_df
```

```python
min_temp = float(input("What is the minimum temperature you want for your trip? "))
max_temp = float(input("What is the maximum temperature you want for your trip? "))
```

```python
preferred_temp_df = city_data_df.loc[(city_data_df["Max Temp"] <= max_temp) & \
                                       (city_data_df["Max Temp"] >= min_temp)]
preferred_temp_df.head(10)
```

```python
preferred_temp_df = preferred_temp_df.dropna()
```

```python
hotel_df = preferred_temp_df[["City", "Country", "Max Temp", "Current Description", "Lat", "Lng"]].copy()
hotel_df["Hotel Name"] = ""
hotel_df.head(10)
```

```python
# 6a. Set parameters to search for hotels with 5000 meters.
params = {
    "radius": 5000,
    "type": "lodging",
    "key": g_key
}

print("Commencing Hotel Match!      ")
print("-----------------------------")


# 6b. Iterate through the hotel DataFrame.
for index, row in hotel_df.iterrows():
    lat = row["Lat"]
    lng = row["Lng"]

    # 6c. Get latitude and longitude from DataFrame
    params["location"] = f"{lat},{lng}"
    
    # 6d. Set up the base URL for the Google Directions API to get JSON data.
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    # 6e. Make request and retrieve the JSON data from the search. 
    hotels = requests.get(base_url, params=params).json()
    
    # 6f. Get the first hotel from the results and store the name, if a hotel isn't found skip the city.
    try:
        hotel_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
    except (IndexError):
        print("Hotel not found... skipping.")
        
print("-----------------------------")
print("Hotel Match Complete!        ")
print("-----------------------------")
```

```python
hotel_df.replace('', np.nan, inplace=True)
hotel_df = hotel_df.dropna()
hotel_df.head(15)
```

```python
#File Export
output_data_file = "WeatherPy_vacation.csv"
hotel_df.to_csv(output_data_file, index_label="City")
```

```python
info_box_template = """
<dl>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Weather Description</dt><dd>{Current Description}</dd>
<dt>Maximum Temp</dt><dd>{Max Temp}</dd>
</dl>
"""
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]

locations = hotel_df[["Lat", "Lng"]]
```
